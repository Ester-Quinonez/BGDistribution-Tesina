```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datos.csv')
df2 = pd.read_csv('datos2.csv')
personas = pd.read_csv('poblacion.csv')
```

```python
an_completo = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
```

```python
df = df.melt(id_vars = ['Country Name','Country Code','Indicator Name','Indicator Code'],var_name = 'anio', value_name = 'value')
```

```python
df2 = df2.melt(id_vars = ['Country Name','Country Code','Indicator Name','Indicator Code'],var_name = 'anio', value_name = 'value_tot')
```

```python
personas = personas.melt(id_vars = ['Country Name','Country Code','Indicator Name','Indicator Code'],var_name = 'anio', value_name = 'cantidad')
```

```python
df = df[df['anio']!= 'Unnamed: 66']
df2 = df2[df2['anio']!= 'Unnamed: 66']
```

```python
df2 = df2[['Country Name','anio','value_tot']]
personas = personas[['Country Name','anio','cantidad']]
```

```python
df['anio']=df.anio.apply(int)
df2['anio']=df2.anio.apply(int)
personas['anio']=personas.anio.apply(int)
```

```python
df=df[df['anio']>= 2000]
df2=df2[df2['anio']>= 2000]
personas=personas[personas['anio']>= 2000]
```

```python
df2 = df2.merge(right = personas, how='left')
```

```python
df2['value'] = df2['value_tot']/df2['cantidad']
```

```python
guatemala = df[df['Country Name']=='Guatemala']
guatemala2 = df2[df2['Country Name']=='Guatemala']
```

```python
dat_guate = guatemala.value
dat_guate2 = guatemala2.value
```

```python
dat_guate = dat_guate.to_numpy()
dat_guate2 = dat_guate2.to_numpy()
```

```python
mexico = df[df['Country Name']=='Mexico']
mexico2 = df2[df2['Country Name']=='Mexico']
```

```python
dat_mex = mexico.value
dat_mex2 = mexico2.value
```

```python
dat_mex = dat_mex.to_numpy()
dat_mex2 = dat_mex2.to_numpy()
```

```python
elsal = df[df['Country Name']=='El Salvador']
elsal2 = df2[df2['Country Name']=='El Salvador']
```

```python
dat_elsal = elsal.value
dat_elsal2 = elsal2.value
```

```python
dat_elsal = dat_elsal.to_numpy()
dat_elsal2 = dat_elsal2.to_numpy()
```

```python
honduras = df[df['Country Name']=='Honduras']
honduras2 = df2[df2['Country Name']=='Honduras']
```

```python
dat_honduras = honduras.value
dat_honduras2 = honduras2.value
```

```python
dat_honduras = dat_honduras.to_numpy()
dat_honduras2 = dat_honduras2.to_numpy()
```

```python
eeuu = df[df['Country Name']=='United States']
eeuu2 = df2[df2['Country Name']=='United States']
```

```python
dat_eeuu = eeuu.value
dat_eeuu2 = eeuu2.value
```

```python
dat_eeuu = dat_eeuu.to_numpy()
dat_eeuu2 = dat_eeuu2.to_numpy()
```

```python
plt.plot(an_completo, dat_guate, label='Guatemala')
plt.plot(an_completo, dat_elsal, label='El Salvador')
plt.plot(an_completo, dat_honduras, label='Honduras')
plt.plot(an_completo, dat_mex, label='MÃ©xico')

plt.legend(fontsize=8, loc='upper left')
plt.savefig('GEHM-PC')
plt.show()
```
